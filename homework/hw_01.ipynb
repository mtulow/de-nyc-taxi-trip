```python
import pandas as pd
import datetime as dt
```

```python
df = pd.read_csv('../notebooks/data/green/2019/01/green_tripdata_2019-01.csv.gz', parse_dates=['lpep_pickup_datetime','lpep_dropoff_datetime'])
df
```

```python
df.rename(columns=dict(
    VendorID='vendorid',
    lpep_pickup_datetime='pickup_datetime',
    lpep_dropoff_datetime='dropoff_datetime',
    RatecodeID='ratecodeid',
    PULocationID='pickup_locationid',
    DOLocationID='dropoff_locationid'
), inplace=True)
```

```python
df
```

```python
df.loc[(df.pickup_datetime>=dt.datetime(2019,1,15,))&(df.dropoff_datetime<dt.datetime(2019,1,15,23,59,59)), :].count()
```

```python
df.groupby(['pickup_datetime','dropoff_datetime']).filter(lambda row: (row.pickup_datetime>=dt.datetime(2019,1,15))&(row.dropoff_datetime<dt.datetime(2019,1,15,23,59,59)))
```

```python
query = """select count(trip_type) from green_taxi_trips as gpu, green_taxi_trips as gdo, green_taxi_trips as g where gpu.pickup_datetime >= to_date('2019-01-15')"""

df.query(query)
```
